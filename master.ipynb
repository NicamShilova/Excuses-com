```
# Effacement de la liste déroulante "Age"
dbutils.widgets.remove("Age")
```

```
# Création de la liste déroulante "Age"
df = spark.sql("select * from age_ben_snds_csv")
vals = [ str(x[0]) for x in df.select("type").orderBy("type").collect() ]
dbutils.widgets.dropdown("Age", vals[0], vals)

query_value = dbutils.widgets.get("Age")
df_value = spark.sql("select distinct code from age_ben_snds_csv where age_ben_snds_csv.type == '" +  query_value + "'")

try:
  age_ben_snds_csv_value = "and age_ben_snds_csv.code == '" + str(df_value.first().code) + "' "
except:
  logging.info('Not found')
  age_ben_snds_csv_value = ""
  
print(age_ben_snds_csv_value)
```

```
# Recherche d'excuses en fonction de l'assuré et du type d'accident
data_query = spark.sql("select age_ben_snds_csv.code Nombre_Tranche_Age_Beneficiaire, age_ben_snds_csv.type Tranche_Age_Beneficiaire, asu_nat_csv.code Nombre_Nature_Assurance, asu_nat_csv.type Nature_Assurance, att_nat_csv.code Nombre_Nature_Accident_du_Travail, att_nat_csv.type Nature_Accident_du_Travail, ben_qlt_cod_csv.code Nombre_Qualite_du_beneficiaire from extrait_csv, age_ben_snds_csv, att_nat_csv, asu_nat_csv, ben_qlt_cod_csv where extrait_csv.AGE_BEN_SNDS == age_ben_snds_csv.code and extrait_csv.ASU_NAT == asu_nat_csv.code and extrait_csv.ATT_NAT == att_nat_csv.code and extrait_csv.BEN_QLT_COD == ben_qlt_cod_csv.code " + str(age_ben_snds_csv_value) + " order by extrait_csv.AGE_BEN_SNDS")

filtrage = data_query.filter("Nombre_Tranche_Age_Beneficiaire != 99 & Nombre_Nature_Assurance != 99 & Nombre_Nature_Accident_du_Travail != 0 & Nombre_Nature_Accident_du_Travail != 9")

display(filtrage.select("*"))
```

```
# Recherche d'excuses en fonction de la situation l'assuré et celle de ses proches
data_query = spark.sql("select age_ben_snds_csv.code Nombre_Tranche_Age_Beneficiaire, age_ben_snds_csv.type Tranche_Age_Beneficiaire, asu_nat_csv.code Nombre_Nature_Assurance, asu_nat_csv.type Nature_Assurance, att_nat_csv.code Nombre_Nature_Accident_du_Travail, att_nat_csv.type Nature_Accident_du_Travail, ben_qlt_cod_csv.code Nombre_Qualite_du_beneficiaire, ben_qlt_cod_csv.type Nature_Qualite_du_Beneficiaire from extrait_csv, age_ben_snds_csv, att_nat_csv, asu_nat_csv, ben_qlt_cod_csv where extrait_csv.AGE_BEN_SNDS == age_ben_snds_csv.code and extrait_csv.ASU_NAT == asu_nat_csv.code and extrait_csv.ATT_NAT == att_nat_csv.code and extrait_csv.BEN_QLT_COD == ben_qlt_cod_csv.code " + str(age_ben_snds_csv_value) + " order by extrait_csv.AGE_BEN_SNDS")

filtrage = data_query.filter("Nombre_Tranche_Age_Beneficiaire != 99 & Nombre_Nature_Assurance != 99 & Nombre_Nature_Accident_du_Travail != 0 & Nombre_Nature_Accident_du_Travail != 9")

display(filtrage.select("*"))
```

```
# Recherche d'excuses en fonction du type detraitemement des patients
data_query = spark.sql("select age_ben_snds_csv.code Nombre_Tranche_Age_Beneficiaire, ben_qlt_cod_csv.code Nombre_Qualite_du_Beneficiaire, exo_mtf_csv.code Nombre_Exoneration_Ticket_Moderateur, exo_mtf_csv.type Motif_Exoneration_Ticket_Moderateur, mdt_typ_cod_csv.code Nombre_Mode_Traitement, mdt_typ_cod_csv.type Mode_Traitement from extrait_csv, age_ben_snds_csv, ben_qlt_cod_csv, exo_mtf_csv, mdt_typ_cod_csv where extrait_csv.AGE_BEN_SNDS == age_ben_snds_csv.code and extrait_csv.BEN_QLT_COD == ben_qlt_cod_csv.code and extrait_csv.EXO_MTF == exo_mtf_csv.code and extrait_csv.MDT_TYP_COD == mdt_typ_cod_csv.code " + str(age_ben_snds_csv_value) + " order by extrait_csv.AGE_BEN_SNDS")

filtrage = data_query.filter("Nombre_Tranche_Age_Beneficiaire != 99 & Nombre_Qualite_du_Beneficiaire != 0 & Nombre_Qualite_du_Beneficiaire != 9 & Nombre_Exoneration_Ticket_Moderateur != 0 & Nombre_Exoneration_Ticket_Moderateur != 99 & nombre_Mode_Traitement != 9")

display(filtrage.select("*"))
```

```
# Recherche d'excuses en fonction de la prestation appliqué et de la nature de l'accident
data_query = spark.sql("select age_ben_snds_csv.code Nombre_Tranche_Age_Beneficiaire, age_ben_snds_csv.type Tranche_Age_Beneficiaire, att_nat_csv.code Nombre_Nature_Accident_du_Travail, att_nat_csv.type Nature_Accident_du_Travail, ben_qlt_cod_csv.code Nombre_Qualite_du_Beneficiaire, ben_qlt_cod_csv.type Nature_Qualite_du_Beneficiaire, ddp_spe_cod_csv.code Nombre_Discipline_Prestation_Executant, ddp_spe_cod_csv.type Discipline_Prestation_Executant from extrait_csv, age_ben_snds_csv, att_nat_csv, ben_qlt_cod_csv, ddp_spe_cod_csv where extrait_csv.AGE_BEN_SNDS == age_ben_snds_csv.code and extrait_csv.ATT_NAT == att_nat_csv.code and extrait_csv.BEN_QLT_COD == ben_qlt_cod_csv.code and extrait_csv.DDP_SPE_COD == ddp_spe_cod_csv.code " + str(age_ben_snds_csv_value) + " order by extrait_csv.AGE_BEN_SNDS")

filtrage = data_query.filter("Nombre_Tranche_Age_Beneficiaire != 99 & Nombre_Nature_Accident_du_Travail != 0 & Nombre_Nature_Accident_du_Travail != 9  & Nombre_Qualite_du_Beneficiaire != 0 & Nombre_Qualite_du_Beneficiaire != 9 & Nombre_Discipline_Prestation_Executant != 0 & Nombre_Discipline_Prestation_Executant != 121")

display(filtrage.select("*"))
```
